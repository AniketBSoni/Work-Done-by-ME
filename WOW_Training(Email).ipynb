```python
import pandas as pd
import os
from glob import glob
import numpy as np
```

```python
f1=glob("report*.csv")[0]
f2=glob("OPS SuperHub learning*.xlsx")[0]
f3=glob("User list with activation*.xlsx")[0]
```

```python
report = pd.read_csv(f1)
trainings= pd.read_excel(f2,sheet_name='All Items_MINOC_new template')
userid = pd.read_excel(f3,skiprows=4)
```

```python
Carrotspot1= report[['Item ID','User ID','EMAIL','Attendance Status','Completion Status ID','Completion Date']]
```

```python
Carrotspot1.rename(columns={'Attendance Status':'Item Types'}, inplace = True)
Carrotspot1['Item Types'] = Carrotspot1['Item Types'].str.replace('-COMPLETE','')
```

```python
Carrotspot1['ItemID_Exist'] = Carrotspot1['Item ID'].isin(trainings['lms_item(1)'])
```

```python
Carrotspot1['EMAIL']=Carrotspot1['EMAIL'].str.upper()
```

```python
userid['User e-mail']=userid['User e-mail'].str.upper()
```

```python
Carrotspot1['EMAIL_Exist'] = Carrotspot1['EMAIL'].isin(userid['User e-mail'])
```

```python
#Carrotspot1.to_excel('Test_email.xlsx')
```

```python
def set_TF(Carrotspot1):
    if(Carrotspot1['ItemID_Exist'] and Carrotspot1['EMAIL_Exist']):
        return "True"
    else:
        return "False"
```

```python
Carrotspot2= Carrotspot1.assign(result1=Carrotspot1.apply(set_TF, axis=1))
```

```python
#Carrotspot2.to_excel('Test_email_final.xlsx')
```

```python
Carrotspot2.head()
```

```python
CS2= Carrotspot2[Carrotspot2['result1']=='True']
```

```python
CS2=CS2.drop(['Item Types','Completion Status ID','Completion Date','ItemID_Exist','EMAIL_Exist'], axis=1)
```
